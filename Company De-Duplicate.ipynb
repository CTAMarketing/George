```python
# get psl_file
psl = publicsuffix.fetch()
psl = PublicSuffixList(psl)

def get_base_domain(url):
    # This causes an HTTP request; if your script is running more than,
    # say, once a day, you'd want to cache it yourself.  Make sure you
    # update frequently, though!
    hostname = urlparse(url).hostname 
    return psl.get_public_suffix(hostname)



```

```python
def check_duplicate(url):
    url = str(url)
    values = all_domain.values()
    #status = url in all_domain.values()
    status = url in [x for v in all_domain.values() for x in v] or url in values
    if status == False:
        return status
    if status == True:
        k_val = [k for k, v in all_domain.items() if url in v]
        return str(k_val[0])

print (df.iloc[1,5])
check_duplicate(df.iloc[1,5])
```

```python
df['Domain Match Status'] = df['Domain Update'].apply(check_duplicate)
df.to_excel('Hosting Company Cleaned Update.xlsx')
```
